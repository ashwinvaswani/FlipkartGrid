```python
import numpy as np
import pandas as pd
import tensorflow as tf
import io
import cv2
from sklearn.model_selection import train_test_split
import glob
from keras.models import Sequential
from keras.layers import Conv2D,Conv3D
from keras.layers import MaxPool2D,AvgPool2D
from keras.layers import Flatten,LeakyReLU
from keras.layers import Dense,Dropout
import keras
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from keras import backend as K
```

```python
df=pd.read_csv('training.csv')
```

```python
df.head()
```

```python
df['combine'] = df[['x1', 'x2','y1','y2']].values.tolist()
df.head()
```

```python
train_img_list = glob.glob("images/*")
```

```python
label=[]
data=[]
i=0
for imagePath in df['image_name'].values:
    imgn='images\\'+ imagePath
    if imgn in train_img_list :
        image = cv2.imread(imgn)
        imgne=np.array(image)
        #imgnew=imgne.flatten().reshape(1,480*640*3)
        image = cv2.resize(image, (40,30))
        blur = cv2.GaussianBlur(image,(5,5),0)
        #gray = cv2.cvtColor(blur, cv2.COLOR_BGR2GRAY)
        #gray=gray.resize(gray,(160,120))
        print(i)
        i=1+i
        data.append(image)
        label.append(imagePath)
```

```python
d=np.array(data)/255
```

```python
d=d.reshape(-1,30,40,1)
```

```python
d.shape
```

```python
Y_train=df.iloc[:,1:5]
```

```python
Y_train=df.iloc[:,1:5]
Y_train['centerx']=(Y_train['x1']+Y_train['x2'])/2
Y_train['centery']=(Y_train['y1']+Y_train['y2'])/2
Y_train['width']=(Y_train['x2']-Y_train['x1'])
Y_train['height']=(Y_train['y2']-Y_train['y1'])
```

```python
Y_train=Y_train.drop(['x1','x2','y1','y2'],axis=1)
```

```python
Y_train.head()
Y=df.iloc[:,:]
```

```python
Y_train.head()
```

```python
def mean_iou(y_true, y_pred):
   score, up_opt = tf.metrics.mean_iou(y_true, y_pred,num_classes=2)
   K.get_session().run(tf.local_variables_initializer())
   with tf.control_dependencies([up_opt]):
       score = tf.identity(score)
   return score
```

```python
def rmse (y_true, y_pred):
    return K.sqrt(K.mean(K.square(y_pred - y_true), axis=-1))
```

```python
(trainX, valX, trainY, valY) = train_test_split(d,Y_train, test_size=0.05)
```

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(3,3), activation='relu',  input_shape=[30,40,3]))
model.add(Conv2D(64, kernel_size=(1,1),activation='relu'))
#model.add(Conv2D(32, kernel_size=(3,3), activation='relu'))
model.add(AvgPool2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
#model.add(Conv2D(3, kernel_size=(1,1), activation='relu'))
#model.add(AvgPool2D(pool_size=(2, 2)))
#model.add(Conv2D(1, kernel_size=(2,2), activation='relu'))
#model.add(Conv2D(1, kernel_size=(3,3), activation='relu'))
#model.add(Dropout(0.4))
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
#model.add(Dropout(0.2))
model.add(Dense(2, activation='relu'))
#model.add(LeakyReLU(0.1))
model.summary()
```

```python
model.compile(loss=keras.losses.MSE,
              optimizer=keras.optimizers.adam(),
              metrics=['accuracy'])
 
model.fit(trainX, trainY.drop(['height','width'],axis=1), batch_size=50,validation_data=(valX, valY.drop(['height','width'],axis=1)) , epochs=16, verbose=1)
```

```python
model_json_cxy1 = model.to_json()
with open("model.json", "w") as json_file:
    json_file.write(model_json_cxy1)
# serialize weights to HDF5
model.save_weights("model_cxy1.h5")
print("Saved model to disk")
```

```python
model1 = Sequential()
model1.add(Conv2D(32, kernel_size=(3,3), activation='relu',  input_shape=[30,40,3]))
model1.add(Conv2D(64, kernel_size=(1,1),activation='relu'))
#model.add(Conv2D(32, kernel_size=(3,3), activation='relu'))
model1.add(MaxPool2D(pool_size=(2, 2)))
model1.add(Dropout(0.25))
#model.add(Conv2D(3, kernel_size=(1,1), activation='relu'))
#model.add(AvgPool2D(pool_size=(2, 2)))
#model.add(Conv2D(1, kernel_size=(2,2), activation='relu'))
#model.add(Conv2D(1, kernel_size=(3,3), activation='relu'))
#model.add(Dropout(0.4))
model1.add(Flatten())
model1.add(Dense(64, activation='relu'))
model1.add(Dense(32, activation='relu'))
#model.add(Dropout(0.2))
model1.add(Dense(2, activation='relu'))
#model.add(LeakyReLU(0.1))
model1.summary()
```
